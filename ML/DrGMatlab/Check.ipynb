```python
import pathlib
cwd = pathlib.Path.cwd()
from PIL import Image
import numpy as np
import os
import statistics
from matplotlib import pyplot as plt
```

```python
drg_file = cwd / "Images" / "Intron" / "Image00001.png"
```

```python
exon_png = Image.open(drg_file)
exon_png2np = np.asarray(exon_png)
print(exon_png2np.shape)
print(exon_png2np)
print(np.sum(exon_png2np))
```

```python
def norm_value(image, norm: bool = False):
    '''
    '''
    png = Image.open(image)
    png = np.asarray(png)
    if norm:
        png = png / 255000

    return np.sum(png)
```

```python
drg_exon_files = cwd / "Images" / "Exon"
drg_intron_files = cwd / "Images" / "Intron"

eas_exon_files = pathlib.Path("E:\\Negative_Control\\Group_1\\Exon")
eas_intron_files = pathlib.Path("E:\\Negative_Control\\Group_1\\Intron")

eas_exon_files_v2 = cwd.parent.parent / "Data_Files" / "Primates" / "Genetics" / "Homo_Sapiens" / "GHM_64_v8_copy" / "Exon"
eas_intron_files_v2 = cwd.parent.parent / "Data_Files" / "Primates" / "Genetics" / "Homo_Sapiens" / "GHM_64_v8_copy" / "Intron"
```

```python
drg_exon_fnorms = list()
drg_exon_tnorms = list()
for root, _, files in os.walk(drg_exon_files):
    for file in files:
        drg_exon_fnorms.append(norm_value(f"{root}/{file}"))
        drg_exon_tnorms.append(norm_value(f"{root}/{file}", norm = True))
```

```python
drg_intron_tnorms = list()
drg_intron_fnorms = list()
for root, _, files in os.walk(drg_intron_files):
    for file in files:
        drg_intron_fnorms.append(norm_value(f"{root}/{file}"))
        drg_intron_tnorms.append(norm_value(f"{root}/{file}", norm = True))
```

```python
drg_exon_fmean = statistics.mean(drg_exon_fnorms)
drg_intron_fmean = statistics.mean(drg_intron_fnorms)

print(f"Mean for exon images: {drg_exon_fmean}")
print(f"Max value: {max(drg_exon_fnorms)}")
print(f"Min value: {min(drg_exon_fnorms)}")

print(f"Mean for intron images: {drg_intron_fmean}")
print(f"Max value: {max(drg_intron_fnorms)}")
print(f"Min value: {min(drg_intron_fnorms)}")
```
