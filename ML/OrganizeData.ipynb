```python
def random_set(folder_path: str, train = 0.8, test = 0.1, valid = 0.1):
    '''
    '''
    png_files = set(folder_path.rglob("*.png"))

    total_files = len(png_files)

    train_set = set()
    test_set = set()
    valid_set = set()

    for _ in range(int(train * total_files)):
        train_set.add(png_files.pop())

    for _ in range(int(test * total_files)):
        test_set.add(png_files.pop())

    for _ in range(len(png_files)):
        valid_set.add(png_files.pop())

    return train_set, test_set, valid_set


def move_files(old_path: set, new_folder: pathlib.Path):
    '''
    '''
    file: pathlib.Path
    for file in old_path:
        # print(str(file), type(str(file)))
        # print(str(new_folder))
        shutil.move(str(file), str(new_folder / file.name))

```

Folders

```python
exon_folder = cwd / "FractalImage" / "Exon"
intron_folder = cwd / "FractalImage" / "Intron"
cds_folder = cwd / "FractalImage" / "CDS"
utr5_folder = cwd / "FractalImage" / "UTR5"
utr3_folder = cwd / "FractalImage" / "UTR3"

train_fold = cwd / "FractalImage" / "train"
test_fold = cwd / "FractalImage" / "test"
val_fold = cwd / "FractalImage" / "val"

exon_train_fold = train_fold / "EXON"
cds_train_fold = train_fold / "CDS"
intron_train_fold = train_fold / "INTRON"
utr5_train_fold = train_fold / "UTR5"
utr3_train_fold = train_fold / "UTR3"

exon_train_fold.mkdir(parents = True, exist_ok = True)
cds_train_fold.mkdir(parents = True, exist_ok = True)
intron_train_fold.mkdir(parents = True, exist_ok = True)
utr5_train_fold.mkdir(parents = True, exist_ok = True)
utr3_train_fold.mkdir(parents = True, exist_ok = True)

exon_test_fold = test_fold / "EXON"
cds_test_fold = test_fold / "CDS"
intron_test_fold = test_fold / "INTRON"
utr5_test_fold = test_fold / "UTR5"
utr3_test_fold = test_fold / "UTR3"

exon_test_fold.mkdir(parents = True, exist_ok = True)
cds_test_fold.mkdir(parents = True, exist_ok = True)
intron_test_fold.mkdir(parents = True, exist_ok = True)
utr5_test_fold.mkdir(parents = True, exist_ok = True)
utr3_test_fold.mkdir(parents = True, exist_ok = True)

exon_val_fold = val_fold / "EXON"
cds_val_fold = val_fold / "CDS"
intron_val_fold = val_fold / "INTRON"
utr5_val_fold = val_fold / "UTR5"
utr3_val_fold = val_fold / "UTR3"

exon_val_fold.mkdir(parents = True, exist_ok = True)
cds_val_fold.mkdir(parents = True, exist_ok = True)
intron_val_fold.mkdir(parents = True, exist_ok = True)
utr5_val_fold.mkdir(parents = True, exist_ok = True)
utr3_val_fold.mkdir(parents = True, exist_ok = True)
```

```python
exon_train, exon_test, exon_val = random_set(exon_folder)
intron_train, intron_test, intron_val = random_set(intron_folder)
cds_train, cds_test, cds_val = random_set(cds_folder)
utr5_train, utr5_test, utr5_val = random_set(utr5_folder)
utr3_train, utr3_test, utr3_val = random_set(utr3_folder)
```

```python
move_files(exon_train, exon_train_fold)
move_files(exon_test, exon_test_fold)
move_files(exon_val, exon_val_fold)

move_files(intron_train, intron_train_fold)
move_files(intron_test, intron_test_fold)
move_files(intron_val, intron_val_fold)

move_files(cds_train, cds_train_fold)
move_files(cds_test, cds_test_fold)
move_files(cds_val, cds_val_fold)

move_files(utr5_train, utr5_train_fold)
move_files(utr5_test, utr5_test_fold)
move_files(utr5_val, utr5_val_fold)

move_files(utr3_train, utr3_train_fold)
move_files(utr3_test, utr3_test_fold)
move_files(utr3_val, utr3_val_fold)
```
